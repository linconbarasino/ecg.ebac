```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Carregar os dados do arquivo CSV em um DataFrame
file_path = 'ECGCvdata (2).csv'
df = pd.read_csv(file_path)

# Visualizar as primeiras linhas do DataFrame
print(df.head())

# Informa√ß√µes gerais sobre o DataFrame
print(df.info())

# Estat√≠sticas descritivas dos dados
print(df.describe())

# Verificar a presen√ßa de valores nulos
print(df.isnull().sum())

# Gr√°fico de barras para contar a quantidade de valores √∫nicos em cada coluna categ√≥rica
for col in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Contagem de Valores √önicos em {col}', fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Contagem', fontsize=14)
    plt.xticks(rotation=45)
    plt.show()

# Gr√°fico de histograma para vari√°veis num√©ricas
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribui√ß√£o de {col}', fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Contagem', fontsize=14)
    plt.show()

# Boxplot para visualizar a distribui√ß√£o e detectar outliers em vari√°veis num√©ricas
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, y=col)
    plt.title(f'Boxplot de {col}', fontsize=16)
    plt.ylabel(col, fontsize=14)
    plt.show()

# Gr√°fico de dispers√£o para analisar poss√≠veis rela√ß√µes entre vari√°veis num√©ricas
sns.pairplot(df.select_dtypes(include=['int64', 'float64']))
plt.suptitle('Gr√°fico de Dispers√£o das Vari√°veis Num√©ricas', fontsize=16, y=1.02)
plt.show()

# Gr√°fico de correla√ß√£o entre vari√°veis num√©ricas
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correla√ß√£o', fontsize=16)
plt.show()

# Construir um modelo de regress√£o linear simples (exemplo)
X = df[['feature1', 'feature2']]  # Substitua 'feature1' e 'feature2' pelas suas vari√°veis
y = df['target']  # Substitua 'target' pela sua vari√°vel alvo

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

# Avalia√ß√£o do modelo
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Apresenta√ß√£o de visualiza√ß√£o de dados adicionais
# Adicione aqui suas visualiza√ß√µes adicionais ou outras an√°lises relevantes

# Exemplo de visualiza√ß√£o: Gr√°fico de dispers√£o dos resultados da regress√£o
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.title('Gr√°fico de Dispers√£o dos Resultados da Regress√£o', fontsize=16)
plt.xlabel('Valor Real', fontsize=14)
plt.ylabel('Valor Previsto', fontsize=14)
plt.show()
```
